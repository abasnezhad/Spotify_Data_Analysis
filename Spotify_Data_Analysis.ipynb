```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('tracks.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe().transpose()
```

```python
# 10 least papular songs
df.sort_values('popularity', ascending=True).head(10)
```

```python
# top 10 most papular songs between 90-95
most_papular= df.query('90<popularity<95', inplace=False).sort_values('popularity', ascending=False)[:10]
most_papular
```

```python
# setting index to release column
df.set_index('release_date', inplace=True)
df.index=pd.to_datetime(df.index)
```

```python
df.head()
```

```python
#artist in the 18th row
df['artists'].iloc[18]
```

```python
#conversting duration from ms to s
df['duration']=df['duration_ms'].apply(lambda x: round(x/1000))
df.drop('duration_ms', inplace=True, axis=1)
```

```python
#correlation plot

corr_df=df.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson')
plt.figure(figsize=(14,8))
heatmap=sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='black')
heatmap.set_title('correlation heatmap between variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)

```

```python
# sample of data
sample_df= df.sample(int(0.004*len(df)))
sample_df
```

```python
# plot for high correlated features, energy and loudness

plt.figure( figsize=(10,6))
sns.regplot(data=sample_df, y= 'loudness', x='energy', color='c').set(title=['loudness vs energy'])
```

```python
#total number of songs in each year
date=pd.to_datetime(df.index)
year=list(date.year)
```

```python
sns.displot(year, discrete=True, aspect=2, height=5, kind= 'hist').set(title='number of songs per year')
```

```python
#duration of songs each year
fig, ax= plt.subplots(figsize=(18,7))
fig=sns.barplot(x=year, y= df.duration, ax= ax, errwidth=False).set(title='year vs duration')
plt.xticks(rotation=90)
```

```python
df2=pd.read_csv('SpotifyFeatures.csv')
df2.head()
```

```python
#duration of songs in different genres
plt.title('duration of songs in different genres')
sns.color_palette('rocket', as_cmap= True)
sns.barplot(y='genre', x='duration_ms', data=df2)
plt.xlabel('duration')
plt.ylabel('gen')
```

```python
# top 5 genres by popularity
sns.set_style('darkgrid')
plt.figure(figsize=(10,5))
famous=df2.sort_values('popularity', ascending=False).head(10)
sns.barplot(y='genre', x='popularity', data=famous).set(title='top 5 genres by popularity')
```
